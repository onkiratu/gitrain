```python
#piping 
```

```python
(
    billboard.melt(id_vars = ['year','artist','track','time', 
                                           'date.entered'], var_name = 'week', value_name = 'rank') 
 .groupby('artist')['rank']
 .mean()
)
```

```python
#or at the end of each line put a backslash
```

```python
(
    billboard\
 .melt(id_vars = ['year','artist','track','time', 
                                           'date.entered'], var_name = 'week', value_name = 'rank') \
 .groupby('artist')['rank']\
 .mean()
)
```

```python
weather = pd.read_csv('C:/Users/kirat/AppData/Local/Temp/Temp4_scipy-2019-pandas-master.zip/scipy-2019-pandas-master/data/weather.csv')
weather.head()
```

```python
#melt
weather_long = weather.melt(id_vars = ['id','year','month','element'])
weather_long
```

```python
#label variable and value
weather_long = weather.melt(id_vars = ['id','year','month','element',], var_name = 'day', value_name = 'temp',)
weather_long.head()
```

```python
#opposite of melt  .  pandas pivot and pandas pivot table
#pivot doesn't handle duplicate values
```

```python
weather_long.pivot_table(index = ['id', 'year', 'month', 'day'], columns = 'element', values = 'temp')
```

```python
#if it prints something weird add .reset_index()
```

```python
weather_long.pivot_table(index = ['id', 'year', 'month', 'day'], columns = 'element', values = 'temp').reset_index()
```
